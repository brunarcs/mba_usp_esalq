```python
import pandas as pd
import numpy as np


from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib.pyplot import figure
```

```python
df = pd.read_excel('Dendrograma.xlsx')
```

```python
#Cálculo distância
Z = linkage(df, 'single')
```

```python
#Visual
plt.title('Dendrograma')
plt.xlabel('perfil')
plt.ylabel('distância (Single)')

dendrogram(Z, labels=df.index)
figure(figsize=(48, 16), dpi=80)
plt.figure()
```
